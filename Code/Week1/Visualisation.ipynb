# Plot bivariate normal distribution

```python
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import numpy as np

sigma = [[5, 3], [3, 5]]
mu = [1, 1]
pc1 = [2, 2]
pc2 = [-1, 1]

x, y = np.mgrid[-4:6:.01, -4:6:.01]
pos = np.dstack((x, y))
rv = multivariate_normal(mu, sigma)
z = rv.pdf(pos)

fig = plt.figure()
ax = fig.add_subplot(111)
ax.contourf(x, y, z)
ax.scatter(1, 1)
ax.annotate("$\mu$", (1.5, 1.1))

ax.arrow(mu[0], mu[1], pc1[0], pc1[1], width = 0.05) 
ax.arrow(mu[0], mu[1], pc2[0], pc2[1], width = 0.05) 
```
